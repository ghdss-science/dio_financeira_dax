"financial_sample"

```python
financial_sample = pd.read_csv('financial_sample.csv')
```

```python
## Verificação de dados sem NAN
financial_sample.isna().sum()
```

```python
## Verificação de dados duplicados
financial_sample.duplicated().sum()
```

```python
# Remover os símbolos de dólar e as vírgulas, e então converter para float
financial_sample['Units_Sold'] = financial_sample['Units_Sold'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
# Manufacturing_Price
financial_sample['Manufacturing_Price'] = financial_sample['Manufacturing_Price'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
## Sale_Price 
financial_sample['Sale_Price'] = financial_sample['Sale_Price'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
# Primeiro, converter todos os valores para strings
financial_sample['Gross_Sales'] = financial_sample['Gross_Sales'].astype(str)

# Em seguida, remover os símbolos de dólar e as vírgulas, e converter para float
financial_sample['Gross_Sales'] = financial_sample['Gross_Sales'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

```

```python
## Discounts
financial_sample['Discounts'] = financial_sample['Discounts'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
## Sales 

# Primeiro, converter todos os valores para strings
financial_sample['Sales'] = financial_sample['Sales'].astype(str)

# Em seguida, remover os símbolos de dólar e as vírgulas, e converter para float
financial_sample['Sales'] = financial_sample['Sales'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
## COGS

# Primeiro, converter todos os valores para strings
financial_sample['COGS'] = financial_sample['COGS'].astype(str)

# Em seguida, remover os símbolos de dólar e as vírgulas, e converter para float
financial_sample['COGS'] = financial_sample['COGS'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
## Profit

# Primeiro, converter todos os valores para string
financial_sample['Profit'] = financial_sample['Profit'].astype(str)

# Remover símbolos de dólar e vírgulas
financial_sample['Profit'] = financial_sample['Profit'].str.replace('$', '', regex=False).str.replace(',', '', regex=False)

# Substituir valores entre parênteses por negativos
financial_sample['Profit'] = financial_sample['Profit'].str.replace(r'\((.*?)\)', r'-\1', regex=True)

# Converter para float
financial_sample['Profit'] = financial_sample['Profit'].astype(float)
```
